```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Experiment-1 :

```python
image = cv2.imread('F:\Lab2-image\wallpaperflare.com_wallpaper (2).jpg')
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
noise = np.random.normal(0, 1, gray_image.shape)
noisy_image = np.uint8(np.clip(gray_image + noise * 30, 0, 255))
```

```python
average_filtered = cv2.blur(noisy_image, (5, 5))
```

```python
gaussian_filtered = cv2.GaussianBlur(noisy_image, (5, 5), 0)
```

```python
median_filtered = cv2.medianBlur(noisy_image, 5)
```

```python
cv2.imshow('Original Image', gray_image)
cv2.imshow('Noisy Image', noisy_image)
cv2.imshow('Average Filter', average_filtered)
cv2.imshow('Gaussian Filter', gaussian_filtered)
cv2.imshow('Median Filter', median_filtered)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def apply_filters(image_path):
    original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if original_image is None:
        print("Error: Could not open or find the image.")
        return

    perwitt_x = cv2.filter2D(original_image, -1, np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]))
    perwitt_y = cv2.filter2D(original_image, -1, np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]))

    sobel_x = cv2.Sobel(original_image, cv2.CV_64F, 1, 0, ksize=3)
    sobel_y = cv2.Sobel(original_image, cv2.CV_64F, 0, 1, ksize=3)

    plt.figure(figsize=(12, 6))

    plt.subplot(231), plt.imshow(original_image, cmap='gray'), plt.title('Original')
    plt.subplot(232), plt.imshow(perwitt_x, cmap='gray'), plt.title('Perwitt X')
    plt.subplot(233), plt.imshow(perwitt_y, cmap='gray'), plt.title('Perwitt Y')

    plt.subplot(234), plt.imshow(original_image, cmap='gray'), plt.title('Original')
    plt.subplot(235), plt.imshow(sobel_x, cmap='gray'), plt.title('Sobel X')
    plt.subplot(236), plt.imshow(sobel_y, cmap='gray'), plt.title('Sobel Y')

    plt.show()
```

```python
apply_filters('F:\Lab2-image\wallpaperflare.com_wallpaper (2).jpg')
```
